## READING THE CSV FILE

```python
df = pd.read_csv("energydata_complete.csv")
df.head()
```

## REMOVING THE DATE AND LIGHTS COLUMNS

```python
df1 = df.drop(['date', 'lights'], axis=1)
df1.head()
```

## NORMALIZE THE DATASET USING THE MINMAXSCALER

```python

# define data
data = np.array(df1)
# print(data)
# # define min max scaler
scaler = MinMaxScaler()
# # transform data
scaled = scaler.fit_transform(data)
df2 = pd.DataFrame(scaled, columns=df1.columns)
df2
```

## Split our dataset into the training and testing dataset

```python
x = df2[["T2"]]
y = df2[["T6"]]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
linear_model = LinearRegression()
#fit the model to the training dataset
linear_model.fit(x_train, y_train)
# predictions
predicted_values = linear_model.predict(x_test)
```

## Question 19
### Train a lasso regression model with an alpha value of 0.001 and obtain the new feature weights with it. How many of the features have non-zero feature weights?

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
y_pred1 = lasso_reg.predict(x_test)
#comparing the effects of regularisation
def get_weights_df(model, feat, col_name):
#this function returns the weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df

lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')
np.count_nonzero(lasso_weights_df['Lasso_weight'])
```

```python
linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')

ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight')

lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')
final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
```

```python
max=final_weights['Linear_Model_Weight'].idxmax()
min=final_weights['Linear_Model_Weight'].idxmin()
final_weights['Features'].iloc[[max,min]] 
```
